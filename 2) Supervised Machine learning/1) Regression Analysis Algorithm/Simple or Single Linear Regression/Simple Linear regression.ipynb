## 1. Into of Linear Regression
<b>Linear Regression</b> is the most basic <a href="https://github.com/deepak2233/Machine-Learning-in-Python">supervised</a> machine learning algorithm. Supervise in the sense that the algorithm can answer your question based on labeled data that you feed to the algorithm. The answer would be like predicting housing prices, Predicting height of a person etc. Here we are going to talk about a regression task using Linear Regression. In the end, we are going to predict housing prices based on the area of the house and many more.

It is one of the most widely known modeling technique. Linear regression is usually among the first few topics which people pick while learning predictive modeling. In this technique, the dependent variable is continuous, independent variable(s) can be <a href="https://github.com/deepak2233/Mathematics-for-Machine-Learning-Data-Science"> continuous or discrete</a>, and nature of regression line is linear. Linear Regression establishes a relationship between dependent variable (Y) and one or more independent variables (X) using a best fit straight line (also known as regression line).

It is represented by an equation <b>Y = mX + b</b>, where a is intercept, b is slope of the line and e is error term. This equation can be used to predict the value of target variable based on given predictor variable(s).

## 4. All Terminology of the Linear Regression

Y= Dependent Variable (Target Variable)<br>
X= Independent Variable (predictor Variable)<br>
a0= intercept of the line (Gives an additional degree of freedom)<br>
a1 = Linear regression coefficient (scale factor to each input value).<br>
ε = random error

- <b>Features</b> - These are the independent variables in any dataset represented by x1 , x2 , x3,
x4,... xn for ‘n’ features or you can say inputs given during model.
- <b>Target / Output Variable/ Lable</b> - This is the dependent variable whose value depends the
independent variable by a relation (given below) and is represented by ‘y’ .
- <b>Function or Hypothesis of Linear Regression is represented by</b> -
y = m1.x1 + m2.x2 + m3.x3 + … + mn.xn + b<br>
<b>Note:</b> Hypothesis is a function that tries to fit the data.
- <b>Intercept</b> - Here b is the intercept of the line. We usually include this ‘b’ in the equation
of ‘m’ and take ‘x’ values for that ‘m’ to be 1. So modified form of above equation is as
follows:
y = mx
Where mx = m1.x1 + m2.x2 + m3.x3 + … + mn.xn + mn+1.xn+1
Here mn+1 is b and xn+1 = 1
- <b>Training Data</b> - This data contains a set of dependent variables that is ‘x’ and a set of
output variable, ‘y’. This data is given to the machine for it to learn or get trained on
some function (here the function is the equation given above) so that in future on giving
some new values of ‘x’ (called testing data), our machine is able to predict values of ‘y’
based on that function.
- <b>Test Data </b>The test set is a set of observations used to evaluate the performance of the model using some performance metric. It is important that no observations from the training set are included in the test set.
- <b>NOTE :</b> Train/Test is a method to measure the accuracy of your model. It is called Train/Test because you split the the data set into two sets: a training set and a testing set. 80% for training, and 20% for testing. You train the model using the training set.
- <b> Residual Error- </b>A residual is the vertical distance between a data point and the regression line. Each data point has one residual. They are positive if they are above the regression line and negative if they are below the regression line. If the regression line actually passes through the point, the residual at that point is zero.

## 5. Types of Linear Regression
Types of Linear Regression
Linear regression can be further divided into two types of the algorithm:

- <b>Simple Linear Regression:</b><br>
If a single independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Simple Linear Regression i.e single X and Single Y.
-<b> Multiple/Multivariable Linear regression:</b><br>
If more than one independent variable is used to predict the value of a numerical dependent variable, then such a Linear Regression algorithm is called Multiple Linear Regression i.e. Single X and Multiple Y.
- <b>Multivariate Regression</b><br>
It is a method used to measure the degree at which more than one independent variable (predictors) and more than one dependent variable (responses), are linearly related i.e. Multiple X and Multiple Y.

 ### Implement Linear regression without Sklearn library
 To implement the we have to find itself slop(m) and intercept points(c)
 - m = $\frac{\sum(x-x')}{\sum(y-y')^2}$ i.e. x' is mean
 - c = y' - (m*x') 

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

## Reda the data
data = pd.read_csv('D:\\Downloads\\data.csv',header = None , names=['Dependent' ,'Independent'])

data
```

```python
## Collecting X and Y
X = data['Dependent'].values
Y = data['Independent'].values
```

```python
# In order to find the value of m and c, you first need to calculate the mean of X and Y

# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)
 
# Total number of values
n = len(X)
 
# Using the formula to calculate m and c
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
m = numer / denom
c = mean_y - (m * mean_x)
 
# Print coefficients
print(m, c)

```

```python
# The value of m and c from above will be added to this equation
# Plotting Values and Regression Line
max_x = np.max(X) + 100
min_x = np.min(X) - 100
# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x 
 
# Ploting Line
plt.plot(x, y, color='#52b920', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='#ef4423', label='Scatter Plot')
 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

## 9. Types of Error Function.
There are three types of Regresion Cost Function
- Mean Error / Average Eroor
- Mean Absolute Error(MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)

<b>Regression cost Function:</b>
Regression models deal with predicting a continuous value for example salary of an employee, price of a car, loan prediction, etc. A cost function used in the regression problem is called “Regression Cost Function”. They are calculated on the distance-based error as follows:

<b>1. Mean Error (ME)</b>
- <b>Error</b> = y-y’

Where,
Y – Actual Input
Y’ – Predicted output

- In this cost function, the error for each training data is calculated and then the mean value of all these errors is derived.
- Calculating the mean of the errors is the simplest and most intuitive way possible.
- The errors can be both negative and positive. So they can cancel each other out during summation giving zero mean error for the model.
- Thus this is not a recommended cost function but it does lay the foundation for other cost functions of regression models.

<b> 2.Mean Absolute Error (MAE)</b>

- This cost function also addresses the shortcoming of mean error differently. Here an absolute difference between the actual and predicted value is calculated to avoid any possibility of negative error.
- So in this cost function, MAE is measured as the average of the sum of absolute differences between predictions and actual observations.

- MAE Cost function
          MAE = (sum of absolute errors)/n

- It is also known as L1 Loss.
- It is robust to outliers thus it will give better results even when our dataset has noise or outliers.

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('D:\\Desktop\datasets\\2.png')
plt.figure(figsize=(5,5))
plt.title("MAE")
plt.axis("off")
imgplot = plt.imshow(img)
```

<b>Mean Squared Error (MSE)</b>

- This improves the drawback we encountered in Mean Error above. Here a square of the difference between the actual and predicted value is calculated to avoid any possibility of negative error.
- It is measured as the average of the sum of squared differences between predictions and actual observations.
- MSE cost function
        MSE = (sum of squared errors)/n

- It is also known as L2 loss.
- In MSE, since each error is squared, it helps to penalize even small deviations in prediction when compared to MAE. But if our dataset has outliers that contribute to larger prediction errors, then squaring this error further will magnify the error many times more and also lead to higher MSE error.
Hence we can say that it is less robust to outliers. <b> Going forword we will use this funtion</b>

<b>Root Mean Squared Error (RMSE)</b>
RMSE is the most popular evaluation metric used in regression problems. It follows an assumption that error are unbiased and follow a normal distribution. Here are the key points to consider on RMSE:

- The power of ‘square root’  empowers this metric to show large number deviations.
- The ‘squared’ nature of this metric helps to deliver more robust results which prevents cancelling the positive and negative error values. In other words, this metric aptly displays the plausible magnitude of error term.
- It avoids the use of absolute error values which is highly undesirable in mathematical calculations.
- When we have more samples, reconstructing the error distribution using RMSE is considered to be more reliable.
- RMSE is highly affected by outlier values. Hence, make sure you’ve removed outliers from your data set prior to using this metric.
- As compared to mean absolute error, RMSE gives higher weightage and punishes large errors.
